```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

* Player Count

```python
# Display the total number of players
player_count = len(purchase_data_df["SN"].unique())
player_count_df = pd.DataFrame({"Total Players":[player_count]})
player_count_df
```

* Purchasing Analysis (Total)

```python
# Sums of all the column into new data frame
unique_items = purchase_data_df['Item ID'].unique()
unique_items_for_df = len(unique_items)

avg_price = purchase_data_df['Price'].mean()
number_purchase = purchase_data_df['Purchase ID'].count()
total_rev = purchase_data_df['Price'].sum()
purchasing_analysis_df = pd.DataFrame({"Unique_Items": [unique_items_for_df], "Average Price": [avg_price],
                                      "Number of Purhcases": [number_purchase], "Total Revenue": [total_rev]})
```

```python
# Formatting
purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:.2f}".format)
purchasing_analysis_df
```

* Gender Demographics

```python
# Create df of Male v Female v Else
no_repeat_sn_df = purchase_data_df.drop_duplicates(subset='SN', keep="first")
gender_df = pd.DataFrame(no_repeat_sn_df["Gender"].value_counts())

# Gender Demographics
gender_df['Percent of Players'] = gender_df['Gender']/gender_df['Gender'].sum()

# Formatting
gender_df['Percent of Players'] = gender_df['Percent of Players'] * 100
gender_df["Percent of Players"] = gender_df["Percent of Players"].map("{:.2f}%".format)
gender_df
```

* Purchasing Analysis by Gender

```python
# Total Count of Gender
gender_purchases = purchase_data_df["Gender"].value_counts()
grouped_gender = purchase_data_df.groupby(['Gender'])
purchases_sum_by_gender = grouped_gender["Price"].sum()

# Average purchase by Gender
average_by_gender = purchases_sum_by_gender / gender_purchases

# Average purchase by Gender per Person
average_by_person = purchases_sum_by_gender / gender_df['Gender']

# Fix Column Names
gender_purchases_df = gender_purchases.rename_axis('Gender').reset_index(name='Purchase Count')
average_by_gender_df = average_by_gender.rename_axis('Gender').reset_index(name='Average Purchase Price')
average_by_person_df = average_by_person.rename_axis('Gender').reset_index(name='Avg Total Purchase Per Person')
purchases_sum_by_gender_df = purchases_sum_by_gender.rename_axis('Gender').reset_index(name='Total Purchase Value')

# Merge Dataframes
gender_summary_1 = pd.merge(gender_purchases_df, average_by_gender_df, on="Gender")
gender_summary_2 = pd.merge(gender_summary_1, purchases_sum_by_gender_df, on="Gender")
gender_summary_final = pd.merge(gender_summary_2, average_by_person_df, on="Gender")

```

```python
# Final Output
gender_final_df = gender_summary_final.set_index("Gender")

# Formatting
gender_final_df["Average Purchase Price"] = gender_final_df["Average Purchase Price"].map("${:.2f}".format)
gender_final_df["Total Purchase Value"] = gender_final_df["Total Purchase Value"].map("${:.2f}".format)
gender_final_df["Avg Total Purchase Per Person"] = gender_final_df["Avg Total Purchase Per Person"].map("${:.2f}".format)
gender_final_df.head()
```

* Age Demographics

```python
# Age Demographics (binning)

bins = [0, 9, 14, 19, 24, 29, 100]
age_range = ["Under 10", "10-14", "15-19", "20-24", "25-29", "Above 30"]
purchase_data_df["Age Groups"] = pd.cut(purchase_data_df["Age"], bins, labels=age_range)
purchase_data_df.head()
```

```python
# Total purchases by age group
total_purchase_age = purchase_data_df["Age Groups"].value_counts()
total_purchase_age

# Grab avg price per purchase
age_group_df = purchase_data_df.groupby("Age Groups").mean()
age_group_avg_price = age_group_df["Price"]

#Total Purchase Price by age group
age_group_sum_df = purchase_data_df.groupby("Age Groups").sum()
age_group_total_price = age_group_sum_df["Price"]

# Avg purchase price by age
avg_price_by_age = age_group_total_price / total_purchase_age
avg_price_by_age
```

```python
#Put the Age Demo into a DF
age_demo_df = pd.DataFrame(total_purchase_age)
age_demo_df["Average Price"] = age_group_avg_price
age_demo_df["Total Purchase Price"] = age_group_total_price
age_demo_df["Average Price per Purchase"] = age_group_total_price
age_demo_df
```

```python
# Formatting
age_demo_df["Average Price"] = age_demo_df["Average Price"].map("${:.2f}".format)
age_demo_df["Total Purchase Price"] = age_demo_df["Total Purchase Price"].map("${:.2f}".format)
age_demo_df["Average Price per Purchase"] = age_demo_df["Average Price per Purchase"].map("${:.2f}".format)
age_demo_df
```

* Most Popular Items

```python
# Number of purchases by Item ID
purchase_count = purchase_data_df["Item ID"].value_counts()
purchase_count_df = pd.DataFrame(purchase_count)
purchase_count_df.head()

# change Item Id title to purchase count
purchase_count_rename = purchase_count_df.rename(columns={"Item ID": "Purchase Count"})
first_five_df = purchase_count_rename.head(5)
first_five_df
```

```python
# Index dataframe by Item ID
item_indexed_data = purchase_data_df.set_index(["Item ID"])
item_indexed_data.head()
```

```python
# Find Item name and Price information for these Item IDs
top_five = item_indexed_data.loc[[178, 82, 108, 145, 92],["Item Name", "Price"]]
top_five_condenced = top_five.iloc[[0, 13, 22, 31, 40],[0, 1]]

# Merge top five and purchase count df
popular_items = top_five_condenced.join(first_five_df, how='outer')
popular_items
```

```python
# set variables
price = popular_items["Price"]
purchase_count = popular_items["Purchase Count"]

# calculate and add total purchase value
total_purchase_value = price * purchase_count
popular_items["Total Purchase Value"] = total_purchase_value
popular_items
```

```python
# Formatting
popular_items["Price"] = popular_items["Price"].map("${:.2f}".format)
popular_items["Total Purchase Value"] = popular_items["Total Purchase Value"].map("${:.2f}".format)
popular_items
```

* Most Profitable Items

```python
# find unique item ID's
no_repeat_itemid_df = purchase_data_df.drop_duplicates(subset='Item ID', keep="first")
no_repeat_itemid_df.head()

# Index dataframe by Item ID
item_index_data = no_repeat_itemid_df.set_index(["Item ID"])
item_index_data.head()

# Pull only the columns we need
item_index_slim_df = item_index_data[["Item Name", "Price",]] 
item_index_slim_df.head()
```

```python
# Join item dataframe with purchase counts
profit_df = item_index_slim_df.join(purchase_count_rename, how='outer')
profit_df.tail()

# Calculate Total Profit
num_of_purchases = profit_df["Purchase Count"]
price_of_item = profit_df["Price"]
total_profit = price_of_item * num_of_purchases
profit_df["Total Profit"] = total_profit
profit_df.head()

# Sort by profit to find 5 highest values
sorted_profit = profit_df.sort_values(["Total Profit"], ascending=False)
top_5_profit = sorted_profit.head(5)
top_5_profit
```

```python
# Formatting
top_5_profit["Price"] = top_5_profit["Price"].map("${:.2f}".format)
top_5_profit["Total Profit"] = top_5_profit["Total Profit"].map("${:.2f}".format)
top_5_profit
```
